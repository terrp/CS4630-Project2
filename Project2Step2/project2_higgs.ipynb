## 0 · Setup & Data Load 

```python
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, MiniBatchKMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score

sns.set_theme(style='whitegrid', palette='muted')
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

print('All imports OK')
```

```python
# ── Load cleaned data (output of clean.py) ────────────────────────────────────
df = pd.read_csv('data/processed/higgs_200k.csv')
print('Loaded from CSV')

LABEL_COL    = 'label'
FEATURE_COLS = [c for c in df.columns if c != LABEL_COL]

X = df[FEATURE_COLS].values   # shape (200000, 28)
y = df[LABEL_COL].values      # 0 = background, 1 = signal

print(f'X shape : {X.shape}')
print(f'y counts: signal={y.sum():,}  background={(y==0).sum():,}')
```

```python
# ── Quick EDA — label balance & feature distributions ─────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(12, 8))

# Label balance
pd.Series(y).value_counts().plot.bar(ax=axes[0], color=['orange','brown'])
axes[0].set_title('Label distribution (0=background, 1=signal)')
axes[0].set_xlabel('Label')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=0)

# Feature variance overview
pd.DataFrame(X, columns=FEATURE_COLS).var().sort_values().plot.barh(ax=axes[1])
axes[1].set_title('Feature variance (scaled)')
axes[1].set_xlabel('Variance')

plt.tight_layout()
plt.show()
```
